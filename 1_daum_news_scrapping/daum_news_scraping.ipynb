```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36'}

url = 'https://news.daum.net/breakingnews/?page={}&regDate=20210811'
result = []
i=0
while True:    
    i += 1
    if i == 4: # 테스트용 브레이크장치
        break;
    print(f'page : {i}')
    res = requests.get(url.format(i), headers=headers) # i로 페이지 조정
    if res.status_code == 200:
        html = bs(res.text, 'html.parser')
        cont = html.find('ul',{'class': 'list_news2'}) # 페이지 내 리스트
        try:
            items = cont.findAll('li') # 범위 내 모든 li(뉴스 기사)
        except Exception as e:
            print(str(e))
            break
        else:
            for item in items:
                tit = item.find('strong',{'class':'tit_thumb'}).a # 기사 제목
                res2 = requests.get(tit['href']) # 기사 본문 접속
                res2.raise_for_status()
                html2 = bs(res2.text, 'html.parser')
                # 제목, 등록일, 요약문, 본문
                title = html2.find('h3',{'class':'tit_view'}).get_text()
                pubdate = html2.find('span',{'class':'num_date'}).get_text()
                summary = html2.find('strong',{'class':'summary_view'})
                content = html2.find('div',{'class':'article_view'}).get_text().strip()
                result.append({
                    'page':i,
                    'title':title,
                    'pubdate':pubdate,
                    'summary':summary,
                    'content':content
                })

# dataframe으로~
pd_data = pd.DataFrame(result, columns=['page','title', 'pubdate','summary','content'])
pd_data
```

```python
pd_data.to_excel('./daum_news_scraping.xls', index=False)
```

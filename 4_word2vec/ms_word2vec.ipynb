```python
import sqlite3
import re
from collections import Counter

conn = sqlite3.connect('../3_tokenizing/kr_scrapping.db')
c = conn.cursor()
c.execute("SELECT news_words FROM news_words WHERE pub_date LIKE '202008%'")

result = c.fetchall()
```

```python
li_result = []
for words in result:
    words = re.sub("[-=+#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]","", str(words))
    words = words.split(', ')
    li_result.append(words)
    # print(words)
    # for word in words:
    #     word = word.strip()
    #     key_words.append(word)
# li_result
```

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences=li_result, vector_size=100, window=5, min_count=5, workers=4, sg=0)
# model
key_word = "문재인"

model_result = model.wv.most_similar(key_word)
print(model_result)
```

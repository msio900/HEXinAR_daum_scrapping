```python
import sqlite3
import re
from collections import Counter

conn = sqlite3.connect('test.db')
c = conn.cursor()
c.execute("SELECT news_words FROM test_words WHERE pub_time LIKE '23:3%'")

news_words = c.fetchall()

key_words = []
for words in news_words:
    words = re.sub("[-=+#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]","", str(words))
    words = words.split(', ')
    for word in words:
        word = word.strip()
        key_words.append(word)

# key_words
cnt_1 = Counter(key_words)
# cnt

conn.close()
sorted(cnt_1.items(), key = lambda item: item[1], reverse=True)
```

```python
import sqlite3
import re
from collections import Counter
import pandas as pd

strDate = '20200801'
endDate = '20210901'

dt_index = pd.date_range(start=strDate, end=endDate, freq='M')
dt_list = dt_index.strftime("%Y%m").tolist()
conn = sqlite3.connect('kr_scrapping.db')
c = conn.cursor()

for i in dt_list:
    c.execute(f"SELECT news_words FROM news_words WHERE pub_date LIKE '{i}%'")

    news_words = c.fetchall()

    key_words = []
    for words in news_words:
        words = re.sub("[-=+#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]","", str(words))
        words = words.split(', ')
        for word in words:
            word = word.strip()
            key_words.append(word)

    # key_words
    cnt_words = Counter(key_words)
    # cnt
    cnt_words = dict(cnt_words)

    df = pd.DataFrame(cnt_words.items(), columns=['word', 'word_count'])
    df.to_csv(f'./monthly_words/{i}_words.csv', index=False)

conn.close()
```

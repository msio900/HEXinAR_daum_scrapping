```python
import sqlite3
import re
from collections import Counter

conn = sqlite3.connect('test.db')
c = conn.cursor()
c.execute("SELECT news_words FROM test_words WHERE pub_time LIKE '23:3%'")

news_words = c.fetchall()

key_words = []
for words in news_words:
    words = re.sub("[-=+#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]","", str(words))
    words = words.split(', ')
    for word in words:
        word = word.strip()
        key_words.append(word)

# key_words
cnt_1 = Counter(key_words)
# cnt

conn.close()
sorted(cnt_1.items(), key = lambda item: item[1], reverse=True)
```
